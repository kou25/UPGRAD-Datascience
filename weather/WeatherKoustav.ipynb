```python
df['Months']=df['Formatted Date'].str.split().str.get(0).str.split('-').str.get(1)
```

```python
df
```

```python
df1=df.groupby(by='Months').mean().reset_index()
```

```python
plt.figure(1,figsize=(20,8))
plt.plot(df1['Months'],df1[['Temperature (C)','Apparent Temperature (C)']])
plt.xlabel('Months')
plt.ylabel('Temp in celcius')
plt.legend(['Temperature','Apparent Temperature'])


plt.figure(2,figsize=(20,4))
plt.subplot(121)
plt.plot(df1['Months'],df1['Humidity'],color='r')
plt.xlabel('Months')
plt.ylabel('Humidity')

plt.subplot(122)
plt.plot(df1['Months'],df1['Pressure (millibars)'],color='g')
plt.xlabel('Months')
plt.ylabel('Pressure')
```
